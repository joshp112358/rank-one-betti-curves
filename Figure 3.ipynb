```python
# imports
import matplotlib.pyplot as plt
from ripser import ripser
import pandas as pd
from complexes import *
import random

#plt size
plt.rcParams['figure.figsize'] = [10,7]
plt.rcParams['figure.dpi'] = 100 
```

```python
# Generate Betti Curves

num_of_trials = 25
size_of_matrix = 80
random.seed(0)
np.random.seed(0)

positive_rank1_B0 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))
positive_rank1_B1 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))
positive_rank1_B2 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))
positive_rank1_B3 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))

for i in range(num_of_trials):
    M = rankify(random_rank_k_mat(k=1, n=size_of_matrix))
    results = ripser(M, distance_matrix=True, maxdim=3, do_cocycles=True)
    [B0, B1, B2, B3] = padded_betti_data(results)
    positive_rank1_B0.loc[i] = B0[1]
    positive_rank1_B1.loc[i] = B1[1]
    positive_rank1_B2.loc[i] = B2[1]
    positive_rank1_B3.loc[i] = B3[1]
    
print("positive done")
    
negative_rank1_B0 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))
negative_rank1_B1 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))
negative_rank1_B2 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))
negative_rank1_B3 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))

for i in range(num_of_trials):
    M = rankify(-random_rank_k_mat(k=1, n=size_of_matrix))
    results = ripser(M, distance_matrix=True, maxdim=3, do_cocycles=True)
    [B0, B1, B2, B3] = padded_betti_data(results)
    negative_rank1_B0.loc[i] = B0[1]
    negative_rank1_B1.loc[i] = B1[1]
    negative_rank1_B2.loc[i] = B2[1]
    negative_rank1_B3.loc[i] = B3[1]

print("negative done")

mix_sign_rank1_B0 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))
mix_sign_rank1_B1 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))
mix_sign_rank1_B2 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))
mix_sign_rank1_B3 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))

for i in range(num_of_trials):
    M = rankify(mixed_sign_rank_k_mat(k=1, n=size_of_matrix))
    results = ripser(M, distance_matrix=True, maxdim=3, do_cocycles=True)
    [B0, B1, B2, B3] = padded_betti_data(results)
    mix_sign_rank1_B0.loc[i] = B0[1]
    mix_sign_rank1_B1.loc[i] = B1[1]
    mix_sign_rank1_B2.loc[i] = B2[1]
    mix_sign_rank1_B3.loc[i] = B3[1]

print("mix done")

uniform_random_B0 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))
uniform_random_B1 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))
uniform_random_B2 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))
uniform_random_B3 = pd.DataFrame(index = range(num_of_trials), columns = range(10001))

for i in range(num_of_trials):
    M = rankify(uniform_random_symmetric(size_of_matrix))
    results = ripser(M, distance_matrix=True, maxdim=3, do_cocycles=True)
    [B0, B1, B2, B3] = padded_betti_data(results)
    uniform_random_B0.loc[i] = B0[1]
    uniform_random_B1.loc[i] = B1[1]
    uniform_random_B2.loc[i] = B2[1]
    uniform_random_B3.loc[i] = B3[1]

print("random done")
```
