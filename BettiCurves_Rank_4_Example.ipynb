```python
import numpy as np
from complexes import *
from numpy.linalg import matrix_rank
import random

plt.rcParams['figure.figsize'] = [8,5]
plt.rcParams['figure.dpi'] = 100 
```

```python
random.seed(52)
np.random.seed(52)
plt.figure(1)
X = random_rank_k_mat(4,5)
M = rankify(X)
results = ripser(M, distance_matrix=True, maxdim=3, do_cocycles=True)
all_betti_plot(results)
plt.title("Betti Curves for Rank " + str(matrix_rank(X)) + " Matrix")
plt.savefig("./figures/Betti_rank4.png")
```
